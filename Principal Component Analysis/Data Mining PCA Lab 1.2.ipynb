```python
#Dimensionality Reduction using Principle Component Analysis (PCA).

from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy.random as rnd
```

```python
iris = datasets.load_iris()
R = np.array(iris.data)
print(R.shape)

df_iris = pd.DataFrame(iris.data,columns = iris.feature_names)
df_iris['target'] = pd.Series(iris.target)
df_iris.head(150)
```

```python
#ignore target for now
X = df_iris.iloc[:,0:4].values
print(X)
print(np.mean(X))
print(np.std(X))
print(X.shape)
```

```python
X_std = StandardScaler().fit_transform(X)
print(np.mean(X_std))
print(np.std(X_std))
```

```python
#create 150*4 to 150*2 (150*4 => 150*3 => 150*2). 
#For 3D plot use 
#(PC1, PC2, PC3)
#(PC2, PC3, PC4)

#For 2D plot use 
#All possible combinations
```

```python
A = rnd.rand(4,4)
print(A)
```

```python
B = np.matmul(X,A)
print(B)
```

```python
B = np.transpose(B)
covariance = np.cov(B)
print(covariance)
```

```python
eigenvalue, eigenvector = np.linalg.eig(covariance)

print(eigenvalue)
print(eigenvector)
```

```python
eigenvec_T = np.transpose(eigenvector)
Y = np.matmul(eigenvec_T,B)
Y = np.transpose(Y)
print(Y)
print(Y.shape)
```

```python
covariance_2 = np.cov(np.transpose(Y))
print(covariance_2)
```

```python
sorted_eign = np.sort(eigenvalue)[::-1]
print(sorted_eign/np.sum(sorted_eign)*100)
```

```python
new_eigenvector = eigenvec_T[0:3]
print(new_eigenvector.shape)
```

```python
Y = np.transpose(Y)
X = np.matmul(new_eigenvector,Y)
new_150_by_3 = np.transpose(X)
print(new_150_by_3.shape)
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot3D(X[:,0], X[:,1], X[:,2])
ax.view_init(20,20)
plt.show()
```
